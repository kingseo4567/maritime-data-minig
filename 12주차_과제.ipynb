```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "survey_results_public.csv"
raw_data = pd.read_csv(file_path)

# 필요한 열 추출
target_data = raw_data[['Age', 'Country', 'LanguageHaveWorkedWith', 'LearnCode']]

# LearnCode 전처리
methods = target_data['LearnCode'].dropna().str.split(';')
exploded_methods = methods.explode()

# 빈도 계산 및 시각화
size_by_methods = exploded_methods.groupby(exploded_methods).size()
size_by_methods.nlargest(10).plot.pie(figsize=(6, 6), autopct='%1.0f%%')
plt.title("Top 10 Ways Developers Learned to Code")
plt.ylabel('')
plt.tight_layout()
plt.show()
```
